```python
#Importando as bibliotecas 
import dlib
import cv2
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
```

```python
#Abrindo a imagem com as pessoas para começar a treinar o modelo 
imagem = cv2.imread("./imagens/photo-people.jpg")
imagem = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
```

```python
#Vendo a imagem 
plt.imshow(imagem)
```

```python
#Convertando a imagem par a escala de cinza para fazer o algoritmo 
imagem_gray = cv2.cvtColor(imagem, cv2.COLOR_RGB2GRAY)
```

```python
#Pegando o classificador dentro da bibliotecas
classificador = cv2.CascadeClassifier("./classificadores/haarcascade_frontalface_default.xml")
```

```python
#Detectando as faces 
faces = classificador.detectMultiScale(imagem_gray, 1.05, 5)
```

```python
#Vendo o total de faces localizadas 
len(faces)
```

```python
#Criando a imagem com as anotações das faces 
imagem_faces = imagem.copy()
```

```python
#Criando as anotações dentro da imagem 
for (x,y,w,h) in faces:
    rectangle = cv2.rectangle(imagem_faces, (x,y), (x+w, y+h), (255,255,0), 4)
    cv2.putText(rectangle, "Face", (x,y-5),cv2.FONT_HERSHEY_SIMPLEX,0.3,(255,127,0),1)
```

```python
#Vendo a imagem 
plt.figure(figsize=(20,10))
plt.imshow(imagem_faces)
```

```python
#Armazenando os rostos encontrados dentro da imagem

#contador 
face_na_image = 0 


for (x,y,w,h) in faces:
    #Pegando onde estão os rostos
    imagem_roi = imagem[y:y+h, x:x+w]

    #Convertando a imagem par a escala de cinza para fazer o algoritmo 
    imagem_roi = cv2.cvtColor(imagem_roi, cv2.COLOR_BGR2RGB)

    #Salvando a imagem
    cv2.imwrite("faces/face_" + str(face_na_image) + ".png", imagem_roi)
    
    face_na_image += 1
```
